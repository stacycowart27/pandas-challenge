```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
school_data_complete["school_name"].unique()
```

```python
# Calculate the total number of unique schools - use Len() on the list of unique schools
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
# Calculate the total number of students - count of the number of student Ids
studentCount = school_data_complete["Student ID"].count()
studentCount 
```

```python
# Look at the shool data dataframe - take a sum of the budget column
school_data
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score - take the mean of the math_score column from school data complete dataframe
avgMathScore = school_data_complete["math_score"].mean()
avgMathScore
```

```python
# Calculate the average (mean) reading score - take the mean of the reading_score column from school data complete dataframe
avgReadingScore = school_data_complete["reading_score"].mean()
avgReadingScore
```

```python
# use conditional to find all students who had a score of 70 or higher in math
passMathStudents = school_data_complete[school_data_complete["math_score"] >= 70]
passMathStudents.head()
numPassMath = len(passMathStudents)
numPassMath
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
# number of students who passed math / total number of students * 100
percentagePassedMath = (numPassMath / studentCount) * 100
percentagePassedMath
```

```python
# use conditional to find all students who had a score of 70 or higher in reading
passReadingStudents = school_data_complete[school_data_complete["reading_score"] >= 70]
passReadingStudents.head()
numPassReading = len(passReadingStudents)
numPassReading
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
# number of students who passed reading / total number of students * 100
percentagePassedReading = (numPassReading / studentCount) * 100
percentagePassedReading
```

```python
# use conditional to find all students who had a score of 70 or higher in math and reading
passingStudents = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
#passingStudents.head()
numPassingStudents = len(passingStudents)
numPassingStudents
```

```python
# Use the following to calculate the percentage of students that passed math and reading
# (number of students who passed math and reading / total number of students * 100
percentagePassed = (numPassingStudents / studentCount) * 100
percentagePassed
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame - make a dictionary of lists
district_summary = pd.DataFrame(
    {
        "Total Schools": [school_count],
        "Total Students": [studentCount],
        "Total Budget": [total_budget],
        "Average Math Score": [avgMathScore],
        "Average Reading Score": [avgReadingScore],
        "% Passing Math": [percentagePassedMath],
        "% Passing Reading": [percentagePassedReading],
        "% Overall Passing": [percentagePassed]
    }

)

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("{:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}%".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:.2f}%".format)

# Display the DataFrame
district_summary
```

```python
# Calculate the average test scores
groupedSchoolDF = school_data_complete.groupby("school_name")
# get the average reading score
avgReadingScoreDF = groupedSchoolDF.mean("reading_score")
# convert to a series by grabbing just the reading_score column
per_school_reading = avgReadingScoreDF["reading_score"]
per_school_reading
```

```python
# Calculate the number of students per school with math scores of 70 or higher
schoolPassingMathDF = school_data_complete[school_data_complete["math_score"] >= 70]
schoolPassingMathDF
```

```python
# use a groupby to show the number of students at each school who passed math
perSchoolPassingMathDF = schoolPassingMathDF.groupby("school_name")
perSchoolPassingMath = perSchoolPassingMathDF["student_name"].count()
perSchoolPassingMath
```

```python
perSchoolPassingDF = studentsPassingDF.groupby("school_name")
perSchoolPassing = perSchoolPassingDF["student_name"].count()
perSchoolPassing
```

```python
# Use the provided code to calculate the passing rates
perSchoolPassingMathPCT = (perSchoolPassingMath / per_school_counts) * 100
perSchoolPassingReadingPCT = (perSchoolPassingReading / per_school_counts) * 100
perSchoolPassingPCT = (perSchoolPassing / per_school_counts) * 100
perSchoolPassingPCT
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame(
    {
        "School Type": school_types1,
        "Total Students": per_school_counts,
        "Total School Budget": perSchoolBudget,
        "Per Student Budget": per_school_capita,
        "Average Math Score": per_school_math,
        "Average Reading Score": per_school_reading,
        "% Passing Math": perSchoolPassingMathPCT,
        "% Passing Reading": perSchoolPassingReadingPCT,
        "% Overall Passing": perSchoolPassingPCT,
        
    }

)

# Formatting
per_school_summary["Total Students"] = per_school_summary["Total Students"].map("{:,}".format)
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary["Average Math Score"] = per_school_summary["Average Math Score"].map("{:.2f}".format)
per_school_summary["Average Reading Score"] = per_school_summary["Average Reading Score"].map("{:.2f}".format)
per_school_summary["% Passing Math"] = per_school_summary["% Passing Math"].map("{:.2f}%".format)
per_school_summary["% Passing Reading"] = per_school_summary["% Passing Reading"].map("{:.2f}%".format)
per_school_summary["% Overall Passing"] = per_school_summary["% Overall Passing"].map("{:.2f}%".format)

# Display the DataFrame
per_school_summary
```
